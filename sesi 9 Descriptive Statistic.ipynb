# Weighted Mean

rata-rata dengan menghitung pembobotan

```python
x = [8, 1, 2.5, 4, 28] #metrics
w = [0.1, 0.2, 0.3, 0.25, 0.15] #Weightages

wmean = sum(w[i] * x[i] for i in range(len(x))) #manual calculation
print(wmean)

wmean = sum(x_ * w_ for (x_, w_) in zip(x, w)) / sum(w)
print(wmean)
```

```python
y, z, w = np.array(x), pd.Series(x), np.array(w)
wmean = np.average(y, weights=w)
print(wmean)
wmean = np.average(z, weights=w)
print(wmean)
```

```python
# Pembuktian manual
def median_function(x):
    n = len(x)
    if n%2 == 1: #Ganjil
        median_ = sorted(x)[round(0.5*(n-1))] #ambil median
    else: #Genap
        x_ord, index = sorted(x) , round(0.5*n)
        median_=0.5 * (x_ord[index-1]+x_ord[index])
    return median_
median_function(x)
```

# Skewness

```python
x = [8.0, 1, 2.5, 4, 28.0]
n = len(x)

mean_ = sum(x) / n
var_ = sum((item - mean_) ** 2 for item in x) / (n - 1)
std_ = var_ ** 0.5

skew_ = (sum((item - mean_)** 3 for item in x)* n / ((n - 1) * (n -2)* std_**3))
skew_
```

# Measures of Correlation Between pairs of data

```python
x = list(range(-10, 11))
y = [0, 2, 2, 2, 2, 3, 3, 6, 7, 4, 7, 6, 6, 9, 4, 5, 5, 10, 11, 12, 14]
x_, y_ = np.array(x), np.array(y)
x__, y__ = pd.Series(x_), pd.Series(y_)
```

# Covariance

```python
n = len(x)
mean_x, mean_y = sum(x) / n, sum(y) / n
cov_xy = (sum((x[k] - mean_x) * (y[k] - mean_y) for k in range(n))/ (n - 1))
cov_xy
```

```python
cov_matrix = np.cov(x_, y_)
cov_matrix
```

```python
x_.var(ddof = 1)
```

```python
y_.var(ddof = 1)
```

```python
cov_xy = cov_matrix[0, 1]
cov_xy
```

```python
cov_xy = cov_matrix[1, 0]
cov_xy
```

```python
cov_xy = x__.cov(y__)
cov_xy
```

# Correlation Coefficient

```python
var_x = sum((item - mean_x)**2 for item in x) / (n - 1)
var_y = sum((item - mean_y)**2 for item in y) / (n - 1)
std_x, std_y = var_x ** 0.5, var_y ** 0.5
r = cov_xy / (std_x * std_y)
r
```

```python
r, p = scipy.stats.pearsonr(x_, y_)
r
```

```python
p
```

```python
corr_matrix = np.corrcoef(x_, y_)
corr_matrix
```

```python
#x-y Plots
x = np.arange(21)
y = 5 + 2 * x + 2 * np.random.randn(21)
slope, intercept, r, *__ = scipy.stats.linregress(x, y)
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='s', label='Data points')
ax.plot(x, intercept + slope * x, label=line)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(facecolor='white')
plt.show()
```

```python
#Heatmaps
matrix = np.cov(x, y).round(decimals=2)
fig, ax = plt.subplots()
ax.imshow(matrix)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('x', 'y'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('x', 'y'))
ax.set_ylim(1.5, -0.5)
for i in range(2):    
    for j in range(2):        
        ax.text(j, i, matrix[i, j], 
        ha='center', va='center', color='w')
plt.show()
```
