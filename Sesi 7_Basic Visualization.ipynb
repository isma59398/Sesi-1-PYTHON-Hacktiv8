```python
#untuk menghapus kolom yang tidak perlu
#axis=0 (merepresentasikan baris) axis=1(merepresentasikan colom)
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head()
```

pie chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie', 
                            figsize=(5, 6), 
                            autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Imigration to canada by Continent [1980-2013]')
plt.axis('equal')
plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', startangle=90, shadow=True, labels=None)

plt.title('Imigration to canada by Continent [1980-2013]', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

membuat piechart persentase imigran baru by continents di tahun 2013

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.05, 0.15]
df_continents['2013'].plot(kind='pie', 
                            figsize=(15, 6), 
                            autopct='%1.1f%%', 
                            startangle=90, 
                            shadow=True, 
                            labels=None, 
                            pctdistance=1.12, 
                            colors=colors_list, 
                            explode=explode_list)
plt.title('Imigration to canada by Continent [1980-2013]', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```
