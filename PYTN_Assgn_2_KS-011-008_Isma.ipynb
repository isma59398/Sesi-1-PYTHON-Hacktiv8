### 7. Confidence Interval

Questions :  Membuat probabilitas GROSS SQUARE FEET pada Wilayah Manhattan

```python
df
```

```python
df[['BOROUGH', 'GROSS SQUARE FEET']]
```

```python
conv_int = df.groupby("BOROUGH").agg({"GROSS SQUARE FEET" : [np.mean, np.std, np.size]})
conv_int
```

```python
mean_tot = conv_int.iloc [1,0]  
sd = conv_int.iloc [1,1]
n = conv_int.iloc [1,2]         
z = 1.96 
se = sd / np.sqrt(n)
print('upper limit of conv_int: ',mean_tot + z* se)
print('lower limit of conv_int: ',mean_tot - z* se)
```

dari data diatas dapat dianalisiskan bahwa, luas mean wilayah pada Manhataan dapat di bagi dua, dimana untuk luas mean wilayah paling besar adalah =  2764.052613556607 .dan luas mean wilayah paling kecil adalah = 2376.656580937509
